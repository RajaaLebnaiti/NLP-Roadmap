### Dealing with pdf files

```python
import PyPDF2
```

```python
# create a pdf file object
data = open('MyDoc.pdf', 'rb')

# create a pdf object reader
reader = PyPDF2.PdfReader(data)

reader
```

```python
# Number of pages
len(reader.pages)
```

```python
# If we want infos of certain page
page = reader.pages[0]
print(page.extract_text())
```

### Tokenization

```python
import nltk
```

#### Line Tokenization

```python
# Tokenization is splitting sentences, text, paragraphs or entire text into small units

data = "Hello ! My name is Rajaa. I am 16 years old. and I am working on ."

tokens = nltk.sent_tokenize(data)
print(tokens)
```

#### Word Tokenization

```python
data = "Hello ! My name is Rajaa."
tokens = nltk.word_tokenize(data)
tokens
```

```python
# This method has certain limit, for example
data = "I don't drink jus !"
tokens = nltk.word_tokenize(data)
tokens
```
