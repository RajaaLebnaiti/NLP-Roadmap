```python
import re
```

```python
texte = "Python est génial"
```

```python
# trouver toutes les lettres minuscules présentes
response = re.findall('[a-z]', texte)
response
```

```python
# extraire tous les nombres présents dans ce texte
texte = "Il y a 123 pommes et 45 oranges. 5"

response = re.findall('[0-9]+', texte)
response
```

```python
# trouver toutes les lettres majuscules dans ce texte
texte = "Mon code promo est AbC123XYz et il expire bientôt."
response = re.findall('[A-Z]+', texte)
response
```

```python
# Python pour trouver tous les emails dans ce texte.
texte = "Mon email est exemple123@test.com et celui de mon ami est ami456@domaine.org."

re.findall('\S+@\S+', texte)
```

```python
# masquer tous les numéros de téléphone en les remplaçant par "XXX-XX-XX-XX-XX"
texte = "Mon numéro de téléphone est 06-12-34-56-78 et celui de mon ami est 07-98-76-54-32."


re.sub('\d', 'x', texte)
```

```python
# transformer cette chaîne en liste
texte = "Python,Java,C++,JavaScript,Go,Rust"
re.split(',', texte)
```

```python
# Trouver tous les chiffres du texte
texte = "Python3 est fun, mais Java8 et C++ sont aussi cool."

re.findall('[0-9]', texte)
```

```python
re.findall('\d', texte)
```

```python
# remplacer toutes les lettres minuscules a, b ou c par X
texte = "aAbBcCabcABC"
re.sub('[abc]', 'x', texte)
```

```python
# trouver toutes les lettres (majuscules et minuscules)
texte = "Python3Java8C++"
re.findall('[A-Za-z]', texte)
```

```python
# remplacer tous les chiffres par #
texte = "a1b2c3A4B5C6"
re.sub('\d', '#', texte)

# ou bien re.sub('[0-9]', '#', texte)
```

```python
re.sub('[0-9]', '#', texte)
```
