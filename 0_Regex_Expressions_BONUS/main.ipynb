```python
import re
```

```python
texte = "Python est génial"
```

```python
# trouver toutes les lettres minuscules présentes
response = re.findall('[a-z]', texte)
response
```

```python
# extraire tous les nombres présents dans ce texte
texte = "Il y a 123 pommes et 45 oranges. 5"

response = re.findall('[0-9]+', texte)
response
```

```python
# trouver toutes les lettres majuscules dans ce texte
texte = "Mon code promo est AbC123XYz et il expire bientôt."
response = re.findall('[A-Z]+', texte)
response
```

```python
# Python pour trouver tous les emails dans ce texte.
texte = "Mon email est exemple123@test.com et celui de mon ami est ami456@domaine.org."

re.findall('\S+@\S+', texte)
```

```python
# trouver tous les caractères qui ne sont pas des lettres minuscules dans ce texte.
texte = "Python3 est fun! 2026."

re.findall('[^a-z]', texte)
```

```python
# remplacer tous les caractères qui ne sont pas des chiffres par *
texte = "Code123 Python456!"
re.sub('[^\d]', '*', texte)
# ou bien 
re.sub('[^0-9]', '*', texte)
```

```python
# Trouver les caractéres qui ne sont pas des lettres majuscules ou miniscules
texte = "PyThOn123!@#"

re.findall('[^A-Za-z]', texte)
```

```python
# remplacer tous les caractères qui ne sont PAS des lettres minuscules par un espace " "
texte = "abcXYZ123***def"
re.sub('[^a-z]', ' ', texte)
```

```python
# trouver toutes les lettres qui ne sont PAS des majuscules.
texte = "AAaaBBbb123"
re.findall('[^A-Z0-9]', texte)
```

```python
# trouver tous les caractères qui ne sont PAS des caractères de mot.
texte = "User_01 a payé 250 euros."
re.findall('[^\w]', texte)
```

```python
# trouver tous les caractères qui sont des espaces (espaces, tabulations, retours à la ligne).
texte = "ID: 45\nCode:\tA9"

re.findall('\s', texte)
```

```python
# rouver tous les caractères qui ne sont PAS des espaces.
texte = "User42\tAdmin\nGuest99"

re.findall('\S', texte)
```

```python
# trouver toutes les séquences de chiffres présentes dans ce texte.
texte = "ID123 ABC 45XYZ 007"
re.findall('\d+', texte)
```

```python
# trouver toutes les séquences contenant la lettre a répétée une ou plusieurs fois.
texte = "aaab aab ab b"
re.findall('[a]+', texte)
```

```python
# trouver toutes les occurrences de color suivies facultativement d’un chiffre.
texte = "color1 color2 colr color3 color20"
re.findall('color\d+', texte)
```

```python
# trouver tous les mots qui commencent par une lettre majuscule et contiennent au moins une lettre ou un chiffre après.
texte = "ID123 start Python3 end42"
re.findall(r'\b[A-Z][A-Za-z0-9]+\b', texte)

```

```python
texte = "Hello Bella yellow"
re.findall('(Hello| yellow)+', texte)
```

```python
# trouver uniquement les emails valides qui se terminent par .com ou .org.

texte = "email test@test.com autre@test.org test@test"

re.findall(r'\b\w+@\w+\.(?:com|org)\b', texte)
```

```python
"""uniquement les références qui :

commencent par Ref

sont suivies d’une lettre majuscule

puis d’un ou plusieurs chiffres

et forment un mot complet
"""
texte = "RefA123 OK RefB45 Ref123 KO"
re.findall( r'\b(?:Ref)[A-Z]+\d+\b', texte)
```

```python
texte = "RefA123 OK RefB45 Ref123 KO RefBE12"
re.findall( r'\b(?:Ref)[A-Z]+\d+\b', texte)  ## suivi de plusieurs lettres majuscule
```

```python
texte = "RefA123 OK RefB45 Ref123 KO RefBE12"
re.findall( r'\b(?:Ref)[A-Z]{1}\d+\b', texte)  # suivi d'une seule lettre majuscule
```

```python
texte = "RefA123 OK RefB45 Ref123 KO RefBE12"
re.findall(r'\bRef[A-Z]\d+\b', texte)
```

```python
"""
Trouver uniquement les éléments qui :

commencent par 4 chiffres

suivis d’un tiret -

suivis de OK

et qui forment un élément complet
"""
texte = "2024-OK 202A-KO 1999-OK 20-OK 32024-OK"
re.findall(r'\b\d{4}-OK\b', texte)
```

```python
""" Trouver uniquement les lignes qui :

commencent par un nom d’utilisateur (lettres uniquement)

peuvent être suivies de chiffres facultatifs (peut etre!)

contiennent un espace

et se terminent par OK

! Chaque élément est sur une ligne différente."""

texte = "user1 OK\nAdmin23 KO\nguest007 OK\nroot OK"
print(texte)
re.findall(r'^[A-Za-z]+\d*\sOK$', texte, re.MULTILINE)  # permet à ^ et $ de fonctionner ligne par ligne
```

```python
""" Trouver uniquement les lignes qui :
commencent par ERROR
suivies d’un espace
suivies de 3 chiffres exactement
et rien d’autre sur la ligne"""


texte = "ERROR 404\nOK 200\nERROR 500\nOK 20\nERROR abc"
re.findall(r'^ERROR\s\d{3}$', texte, re.MULTILINE)
```

```python
texte = "<p>Hello <b>world</b></p>"

re.findall(r'<[^>]*>', texte)
```

```python
texte = "<p>Hello <b>world</b></p>"

re.sub(r'<[^>]*>', '', texte)
```

#### Exercice complet: 

```python
texte = """
Contactez-nous à support@test.com ou admin@mail.org.
Visitez https://example.com ou http://test-site.net pour plus d'infos.
Suivez-nous sur Twitter avec #Python3 et #Regex_101.
Les faux exemples : test@com, http:/site.com, #123, #.
"""
print(texte)
```

Trouver uniquement les emails valides dans le texte.

📌 Contraintes :

format classique nom@domaine.extension

extensions valides : .com, .org

email = mot complet

```python
re.findall(r'\b\w+@\w+\.(?:com|org)\b', texte)
```

Trouver uniquement les URLs valides.

📌 Contraintes :

commencent par http:// ou https://

domaine valide (lettres + tirets)

extension : .com ou .net

URL = mot complet

```python
re.findall(r'\b(?:http://|https://)\S+\.(?:com|net)\b', texte)
```

Trouver uniquement les hashtags valides.

📌 Contraintes :

commencent par #

suivis d’au moins une lettre

peuvent contenir lettres, chiffres, _

pas uniquement des chiffres

hashtag = mot complet

```python
re.findall(r'#[A-Za-z]+\w*\b', texte)
```
