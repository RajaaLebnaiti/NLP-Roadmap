```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
```

```python
#  Generate data

np.random.seed(42)
torch.manual_seed(42)
```

```python
t = np.linspace(0, 100, 100)  # start = 0, end = 100, len(t) = 100 

data = np.sin(t)



def create_sequence(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:(i + seq_length)]
        y = data[i + seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)\
    

seq_length = 10
X, y = create_sequence(data, seq_length)


trainX = torch.tensor(X[:, :, None], dtype= torch.float32)
trainY = torch.tensor(y[:, None], dtype = torch.float32)


```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first= True)
        self.fc = nn.Linear(hidden_size, output_size)
        # self.sigmoid = nn.Sigmoid() # on prédit des valeurs numériques !!!!
        
    
    def forward(self, x):
        h0 = torch.zeros(1, x.size(0), self.hidden_size)
        c0 = torch.zeros(1, x.size(0), self.hidden_size)
        
        out, (h_n, c_n) = self.lstm(x, (h0, c0))
        prediction = self.fc(h_n.squeeze(0))
        return prediction
```

```python
model = LSTMModel(1, 100, 1)

# 1: X a une forme (90, 10).   X[:, :, None] transforme cela en (90, 10, 1) input_size est donc bien 1 (puisqu'il y a 1 valeur à chaque étape temporelle)
# 100 choix personnels 
# 1: output size c'est 1 
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
```

```python
epochs = 100
for epoch in range(epochs):
    model.train()
    y_pred = model(trainX)
    loss = criterion(y_pred ,trainY)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
```
