```python
import torch
import torch.nn as nn

class NLP_GRU(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_size, n_layers):
        super().__init__()
        
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.gru = nn.GRU(embedding_dim, 
                          hidden_dim, 
                          num_layers=n_layers, 
                          batch_first=True, 
                          )
        
        self.fc = nn.Linear(hidden_dim, output_size)
        
    def forward(self, x):
        
        embedded = self.embedding(x)

        output, hn = self.gru(embedded)
        

        
        return self.fc(hn[-1])
```

```python
model = NLP_GRU(vocab_size=8000, embedding_dim=100, hidden_dim=256, output_size=3, n_layers=2)


input_phrase = torch.LongTensor([[4, 150, 89, 2045]]) 


model.eval() 
with torch.no_grad():
    predictions = model(input_phrase)

print(predictions) 
```
