#### Regex Tokenizer

```python
from nltk.tokenize import RegexpTokenizer

text = 'Hello world! I\' am Rajaa '

tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text)
tokens
```

### Stemming

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

words = ['Like', 'Liking', 'Likes', 'Studying', 'Working', 'CLEANED', 'BirDS']

for word in words:
    print(word, " : ", ps.stem(word))
```

### StopWords

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "Data science is one of the most trendind field to work with."

# tokenisation du texte (en miniscule pour uniformer )
tokens = word_tokenize(text.lower())


# récupérer la liste des stop words anglais
stop_words = set(stopwords.words('english'))  # set pour ne pas avoir de répétition

# récupérer les mots qui ne sont pas des stops words
filtred_tokens = [word for word in tokens if word not in stop_words]

# récupérer la liste des stops words
stopwords_tokens = [word for word in tokens if word in stop_words]


print("La liste initiale : ", tokens, "| taille : ", len(tokens))
print("La liste des mots qui ne sont pas des stopwords : ", filtred_tokens, "| taille : ", len(filtred_tokens))
print("La liste des mots qui sont des stopwords : ", stopwords_tokens, "| taille : ", len(stopwords_tokens))
```

#### Les accents


```python
import unicodedata
def remove_accents(text):
    print("text initial : ", text)
    # décompposer les caractéres accentués
    text = unicodedata.normalize('NFD', text)  # é devient e + ´
    print("text normalisé : ", text)
    
    # garder que le caractéres qui ne sont pas des signe diacritiques
    text = "".join([c for c in text if unicodedata.category(c) != 'Mn']) # Parcours chaque morceau, et ne garde que ceux qui ne sont PAS des accents ('Mn') Puis on recolle les morceaux par join
    print("Après : ", text)

remove_accents("Héhé, l'été arrive bientôt !")
```

H et e sont des lettres (Catégorie 'L').

´ (l'accent seul) appartient à la catégorie 'Mn' (Mark, Nonspacing). Ce sont des marques qui ne prennent pas de place horizontale.

#### La ponctuation

```python
import string
print(string.punctuation)
print(str.maketrans('', '', string.punctuation))
```

La fonction maketrans crée une "table de correspondance". Elle prend 3 arguments :
- 1. Les caractères à remplacer.
- 2. Par quoi les remplacer.
- 3. Les caractères à supprimer.

Prépare une table qui ne remplace rien, mais qui supprime absolument tous les signes contenus dans string.punctuation (!, ?, ., ,, etc.)

```python
import string

def remove_punctuation(text):
    # On crée une table de correspondance qui remplace chaque ponctuation par 'None'
    translator = str.maketrans('', '', string.punctuation) 
    return text.translate(translator)

print(remove_punctuation("Bonjour, comment ça va ?")) 
```

```python
text = 'Hello -$ World ??! '
print(text)

translator = str.maketrans('e', 'Z', 'W')
print(translator)

text = text.translate(translator)
print(text)
```

#### Les chiffres

```python
import re

def remove_numbers(text):
    # \d+ correspond à n'importe quel chiffre de 0 à 9
    return re.sub(r'\d+', '', text)

print(remove_numbers("J'ai 25 ans et 2 chats.")) 
```
