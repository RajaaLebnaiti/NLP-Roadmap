# Data Cleaning

```python
data = pd.read_csv('IMDB_clean_Dataset.csv', encoding="utf-8")
data.head()
```

## 1)- Remove html code

```python
import re

def remove_noisy_data(text):
    clean_text = re.sub(r'<[^>]*>', '' ,text)
    return clean_text



data['review'] = data['review'].apply(remove_noisy_data)    
```

```python
data.head()
```

## 2)- Remove spaces

```python
def remove_spaces(text):
    text = text.strip()
    clean_text = re.sub(r'\s+', ' ', text)
    return clean_text


data['review'] = data['review'].apply(remove_spaces)    
```

## 3)- Remove punctuation

```python
import string

def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

data['review'] = data['review'].apply(remove_punctuation)
```

## 4)- Remove numbers

```python
def remove_numbers(text):
    clean_text = re.sub('\d+', '', text)
    return clean_text

data['review'] = data['review'].apply(remove_numbers)

```

## 5)- Remove accents




```python
import unicodedata
def remove_accents(text):
    text= unicodedata.normalize('NFD', text)
    text = "".join([c for c in text if unicodedata.category(c)!= 'Mn'])
    return text

data['review'] = data['review'].apply(remove_accents)
```

## 6)- Lower case 


```python
def lower_case(text):
    text_clean= text.lower()
    return text_clean

data['review'] = data['review'].apply(lower_case)
```

## 7)- Lemmatizing

```python
from nltk.stem import WordNetLemmatizer

def stemming(text):
    lema = WordNetLemmatizer()
    text = ' '.join([lema.lemmatize(word) for word in text.split()])
    return text
data['review']=data['review'].apply(stemming)
```

## 8)- tokenization and remove stop words

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def remove_english_stop_words(text):
    english_stops = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    filtered_tokens = ' '.join([word for word in tokens if word not in english_stops])
    return filtered_tokens


data['review'] = data['review'].apply(remove_english_stop_words)
   
```
