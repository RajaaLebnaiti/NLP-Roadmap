## 1)- Remove noisy data (html code)

```python
from bs4 import BeautifulSoup
import re

def remove_noisy_data(text):
    soup = BeautifulSoup(text, "html.parser")
    text = soup.get_text()
    text = re.sub('\[[^]]*\]', '', text)
    return text



data['review'] = data['review'].apply(remove_noisy_data)    
```
