### 🟢 PARTIE B — Nettoyage du texte (PHASE 2)


Pour chaque texte :

#### 5. Appliquer le lowercasing.



```python
# 5. Pour chaque texte appliquer lowercasing
def lower_case(corpus):
    corpus_final = [] # pour stocker le texte nettoyé
    for doc in corpus:
        doc = doc.lower()
        corpus_final.append(doc)
    return corpus_final

```

#### 6. Supprimer la ponctuation.



```python
# 6. Supprimer la ponctuation
import string
def delete_puntuation(corpus):
    corpus_final = [] # pour stocker le texte nettoyé
    translator = str.maketrans('', '', string.punctuation)
    for doc in corpus:
        doc = doc.translate(translator)
        corpus_final.append(doc)
    return corpus_final

```

#### 7. Supprimer les chiffres.



```python
# 7. supprimer les chiffres
import re
def delete_numbers(corpus):
    corpus_final = [] # pour stocker le texte nettoyé
    for doc in corpus:
        doc = re.sub(r'\d+', '', doc)
        corpus_final.append(doc)
    return corpus_final



```

#### 8. Vérifier que l’encodage est correct (UTF-8).



```python
def encode_utf_8(corpus):
    for doc in corpus:
        try:
            doc.encode('utf-8').decode('utf-8')
            print("L'encodage est OK")
        except UnicodeError:
            print("Probléme d'encodage")
            
encode_utf_8(corpus)
```

### 🟢 PARTIE C — Tokenisation



#### 10. Tokeniser chaque texte mot par mot avec nltk.word_tokenize.



```python
# Tokenizer chaque mot du texte
import nltk


def word_tokenize_corpus(corpus):
    tokens = dict() # enregistrer dans un dictionnaire les tokens ou la clé est l'index du corpus (phrase 1 ou 2 , etc)
    for index, doc in enumerate(corpus):
        doc = nltk.word_tokenize(doc)
        tokens[index] = doc

    return tokens

print(word_tokenize_corpus(corpus))
```

#### 13. Identifier les tokens de ponctuation restants (s’il y en a).

```python
import string
# Identifier les tokens de ponctuation avant de faire la suppression de la ponctuation!
def count_punctuation_tokens(corpus):
    
    punctuations = list(string.punctuation) # on stocke la liste des pontuations
    tokens = word_tokenize_corpus(corpus) # on récupére les tokens
    
    
    punctuation_tokens = dict() # stocker le résultat final
    for key, value in tokens.items():
        punctuation = []
        for v in value:
            if v in punctuations:
                punctuation.append(v)
        punctuation_tokens[key] = punctuation
    return punctuation_tokens
count_punctuation_tokens(corpus)
```

#### 14. Charger la liste des stop words anglais avec NLTK.


#### 15. Supprimer les stop words de chaque document.


```python
# Supprimer les stop words de chaque document
def delete_stopwords(corpus):
    # dictionnaire des tokens
    tokens = word_tokenize_corpus(corpus)
    # La liste des stopswords anglais
    english_stopwords = set(stopwords.words('english'))

    final_corpus = dict() # pour stocker le résultat final
    
    for key, value in tokens.items():
        filtered_tokens = [word for word in value if word not in english_stopwords]
        final_corpus[key] = filtered_tokens
   
    return final_corpus

```

#### 44. Extraire les POS tags des tokens.


```python
import nltk

def pos_tags_extraction(corpus):
    tokens = word_tokenize_corpus(corpus)
    pos_tags = dict()
    
    for idx, token_list in tokens.items():
        pos_tags[idx] = nltk.pos_tag(token_list)
    
    return pos_tags
pos_tags_extraction(corpus)
```


#### 45. Identifier :

    - noms

    - adjectifs

    - verbes


```python
# NN NNS nom

# VB VBD verbs

# JJ adjectives 
# ......

def extract_pos_categories(corpus):
    pos_tags = pos_tags_extraction(corpus)
    
    nouns, verbs, adjectives = [], [], []
    
    for tagged_tokens in pos_tags.values():
        for word, tag in tagged_tokens:
            if tag.startswith('NN'):
                nouns.append(word)
            elif tag.startswith('VB'):
                verbs.append(word)
            elif tag.startswith('JJ'):
                adjectives.append(word)
    
    return nouns, verbs, adjectives
extract_pos_categories(corpus)
```
