## 1- Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "I love Data Science !",
    "Data Science is great."
]

vectorizer = CountVectorizer()

x = vectorizer.fit_transform(corpus)

print("Vocabulaire : ", vectorizer.get_feature_names_out())
print("Matrice BoW :\n", x.toarray())
```

```python
# les limites de cette méthode

# a)- Deux phrases avec le meme vecteur mais le sens est different

corpus = [
    "Le chat mange la souris",
    "La souris mange le chat"
    ]

vectorizer = CountVectorizer()
x= vectorizer.fit_transform(corpus)

print("Vocabulaire est : ", vectorizer.get_feature_names_out())
print("La matrice BoW :\n", x.toarray())
```

```python
# b)- Le cas ou on a des mots composés: ne pas

corpus = [
    "Ce film est bien, il n'est pas mauvais",
    "Ce film est mauvais, il n'est pas bien"
    ]


vectorizer = CountVectorizer()
x= vectorizer.fit_transform(corpus)

print("Vocabulaire est : ", vectorizer.get_feature_names_out())
print("La matrice BoW :\n", x.toarray())
# Meme vecteurs -- meme sens pour la machine (ce qui est faux les sens sont différents)
```

```python
# On peut résoudre ces problémes avec N-Grams
```

## 2- N-Grams

#### a- 1-gram

```python
# 1 gram c'est Bow Classique
```

#### b- 2-gram

```python
# prend deux mots 

corpus = ["L'intellgence artificielle est fascinante", "L'intellgence artificielle"]

# avec 1-gram
vectorizer_1gram = CountVectorizer()

# avec 2-gram
vectorizer_2gram = CountVectorizer(ngram_range=(1,2)) # mélange de unigram et bigram

x1= vectorizer_1gram.fit_transform(corpus)
x2 = vectorizer_2gram.fit_transform(corpus)


print("AVEC 1-GRAM :")
print("Vocabulaire est : | ", vectorizer_1gram.get_feature_names_out())
print("La matrice BoW :\n", x1.toarray())
print("----"*20)
print("AVEC 2-GRAM :")
print("Vocabulaire est : | ", vectorizer_2gram.get_feature_names_out())
print("La matrice BoW :\n", x2.toarray())
```

<b>Limite : </b> 

Explosion de la taille : Si on a 10 000 mots dans le dictionnaire, le nombre de Bigrams possibles est de $10\,000^2$ (100 millions !). L'ordinateur risque de manquer de mémoire.

## 3- TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "Le chat mange",
    "Le chien dort"
]

# 1. Initialisation
vectorizer = TfidfVectorizer()

# 2. Calcul (Fit) et Transformation
tfidf_matrix = vectorizer.fit_transform(corpus)

# 3. Affichage du vocabulaire et des scores
print("Mots :", vectorizer.get_feature_names_out())
print("Matrice TF-IDF :\n", tfidf_matrix.toarray())
```
