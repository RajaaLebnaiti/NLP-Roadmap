# Word2Vec

```python
from gensim.models import Word2Vec
```

```python
sentences = [
                ['L', 'intelligence', 'artificielle', 'est', 'fascinante'],
                ['Le', 'machine', 'learning', 'est', 'un', 'outil', 'génial']
]
```

```python
model = Word2Vec(sentences, window= 5, sg=1, vector_size= 12, min_count= 1)
```

```python
model.wv['intelligence']
```

```python
model.wv.most_similar('intelligence')
```

# Glove

```python
import gensim.downloader as api

model_glove = api.load('glove-wiki-gigaword-100')
```

```python
model_glove
```

```python
# trouver les similarités
model_glove.similarity('apple', 'orange')
```

```python
# analogie roi - homme + femme = ?
model_glove.most_similar(positive=['king', 'woman'], negative=['man'], topn=1)
```

```python
# trouver l'intrus
model_glove.doesnt_match(["fire", "water", "earth", "Book"])
```
